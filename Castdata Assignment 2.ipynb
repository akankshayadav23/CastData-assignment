```python
import pandas as pd
```

```python
def read_data():
    data = pd.read_csv("Downloads/cast.csv")
    return data
```

```python
data = read_data()
print(data)
```

```python
data.head(10)
```

# What are the ten most common movie of all time

```python
movie = data['title'].value_counts()
top_10 = movie.head(10)
print("the 10 most common movies of all the time are")
print(top_10)
```

# Which three years of the 1930s saw the most films released?

```python
years = data[(data['year']>=1930)&(data['year']<=1939)]
top_year = years['year'].value_counts()
top_3_years = top_year.head(3)
print("the three years which saw the most released films in 1930s are:")
print(top_3_years)
```

# plot the number of films that have been released each decade over the history of cinema

```python
data['decade'] = data.year//10*10
```

```python
data.groupby('decade')['n'].mean().plot(kind='bar')
```

# Plot the number of "Hamlet" films made each decade?

```python
hamlet = data[data['title']=='Hamlet']
hamlet_counts = hamlet.groupby('decade')['title'].count()
hamlet_counts.plot(kind='bar',title='number of Hamlet movie each decade')
```

# Plot the number of "Rustler" character in each decade of the history of films?

```python
rus = data[data['character']=='Rustler']
rus_cnt = rus.groupby('decade')['character'].count()
rus_cnt.plot(kind='bar',title='number of rustler character each decade')
```

# PLot the number of "Hamlet" character each decade.

```python
ham = data[data['character']=='Hamlet']
ham_cnt = ham.groupby('decade')['character'].count()
ham_cnt.plot(kind='bar',title='number of Hamlet character each decade')
```

# What are the 11 most common character names in movie history?

```python
chtr = data['character'].value_counts()
chtr_11 = chtr.head(11)
print("the 11 most common character names in movie history are:")
print(chtr_11)
```

# Who are the 10 people most often credited as "Herself" in film history?

```python
hr_chr = data[data['character']=='Herself']
hr_count = hr_chr['name'].value_counts()
top10_herself = hr_count.head(10)
print("the 10 people who are the most often credited as Herself are:")
print(top10_herself)
```

# Which actors or actresses appeared in the most movies in the year 1945?

```python
year = data[data['year']==1945]
actor = year['name'].value_counts()
most_appeared = actor.idxmax()
print("the most appeared actor/actress in the year 1945 is:",most_appeared)
```

# Which actors or actresses appeared in the most movies in the year 1985?

# How many leading(n=1) roles were available to actors , and how many to actresses , in the 1950s?

```python
year = data[(data['year']>=1950)&(data['year']<=1959)&(data['n']==1)]
lead1 = year['type'].value_counts()
lead_actor = lead1.get('actor',0)
lead_actress = lead1.get('actress',0)
print("total leading actors who got roles n=1 was :",lead_actor)
print("total leading actress who got roles n=1 was :",lead_actress)
```
